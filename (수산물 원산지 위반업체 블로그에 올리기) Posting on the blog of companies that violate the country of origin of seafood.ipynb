```python
'''
HTML Parsing: Parse the HTML content of the page using 
BeautifulSoup or another HTML parsing library.'''
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
import requests

url = "http://www.nfqs.go.kr/hpmg/data/actionAnnounceViolatOriginForm.do?menuId=M0000232"

# Initialize the web driver
driver=webdriver.Chrome()

# Navigate to the initial page
driver.get(url)

col1=[]
col2=[]
num_page=2

while True:
    # Get the current page source
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    '''locate the specific table you want to scrape
    you can find the name 'board-list-table' in html '''
    table = soup.find('table', class_='board-list-table')
    #tr :Table Row / td : table Data
    '''
    <table>
    <tr>
        <td>Row 1, Cell 1</td>
        <td>Row 1, Cell 2</td>'''

    rows=table.find_all('tr')

    
    for i in range(len(rows)): 
        if i%2==1 and i>2:
            row = rows[i]
            origin=row.find('td', colspan="3")
            origin =origin.get_text(strip=True)
            if '일본'in origin:
                tem=rows[i-1].find_all('td')
                tem2=row.find_all('td')
                col1.append([tem[0].text.strip(),tem[5].text.strip(),tem[2].text.strip(),tem2[0].text.strip(),tem[4].text.strip(),tem2[1].text.strip(),tem2[2].text.strip(),'일본'])
            elif '러시아'in origin:
                tem=rows[i-1].find_all('td')
                tem2=row.find_all('td')
                col1.append([tem[0].text.strip(),tem[5].text.strip(),tem[2].text.strip(),tem2[0].text.strip(),tem[4].text.strip(),tem2[1].text.strip(),tem2[2].text.strip(),'러시아'])
            elif '중국'in origin:
                tem=rows[i-1].find_all('td')
                tem2=row.find_all('td')
                col1.append([tem[0].text.strip(),tem[5].text.strip(),tem[2].text.strip(),tem2[0].text.strip(),tem[4].text.strip(),tem2[1].text.strip(),tem2[2].text.strip(),'중국'])
    try:
        driver.execute_script("document.querySelector('a.nextpage').click()")
    except:
        break
    # Add a delay to allow the next page to load (adjust the delay time as needed)
    time.sleep(2)
print(col1)
```

```python
df = pd.DataFrame(col1, columns=['등록번호', '날짜', '이름', '주소','어종','구분', '위반내용','비고'])
df
```

```python
import pyperclip
import json
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains

url2 = "https://blog.naver.com/GoBlogWrite.naver"

# Initialize the web driver
driver=webdriver.Chrome()

# Navigate to the initial page
driver.get(url2)

# 그냥입력하면, 자동입력방지 테스트가 뜬다. 클립보드를 이용해 복사.
username_field = driver.find_element(By.ID,"id")  # 사용자 이름 필드의 ID로 변경
username_field.click()
pyperclip.copy(username)
username_field.send_keys(Keys.CONTROL, "v")  # 텍스트 복사
time.sleep(1)

password_field = driver.find_element(By.ID,"pw")  # 사용자 이름 필드의 ID로 변경
password_field.click()
pyperclip.copy(password)
password_field.send_keys(Keys.CONTROL, "v")  # 텍스트 복사
time.sleep(1)

driver.find_element(By.ID,'log.login').click()
time.sleep(2)
```
